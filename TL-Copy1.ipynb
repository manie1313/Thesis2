```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.io



def quadrant_determine(positions):
    quadrant = []
    for pos in positions:
        if pos[0]<0.5 and pos[1]<0.5:
            quadrant.append(1)
        elif pos[0]>0.5 and pos[1]<0.5:
            quadrant.append(2)
        elif pos[0]<0.5 and pos[1]>0.5:
            quadrant.append(3)
        else:
            quadrant.append(4)
    return np.array(quadrant)
            
            
    
    

dic_result = scipy.io.loadmat("Desktop/exploreSLproperties-master/results/Elyes/experiment_1.mat")

duration = dic_result["task_duration"][0]
positionT = dic_result["stimuli_history"][:,0,2:]
quadrantT = quadrant_determine(positionT)
plt.figure()

plt.plot(np.where(quadrantT==1)[0],duration[quadrantT==1],'*',markersize=10)
plt.plot(np.where(quadrantT==2)[0],duration[quadrantT==2],'x',markersize=10)
plt.plot(np.where(quadrantT==3)[0],duration[quadrantT==3],'+',markersize=10)
plt.plot(np.where(quadrantT==4)[0],duration[quadrantT==4],'p',markersize=10)
plt.plot(duration,'--')
plt.ylabel("time (s)")
plt.xlabel("# epoch")
plt.show()

```

```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
import scipy.io



def quadrant_determine(positions):
    quadrant = []
    for pos in positions:
        if pos[0]<0.5 and pos[1]<0.5:
            quadrant.append(1)
        elif pos[0]>0.5 and pos[1]<0.5:
            quadrant.append(2)
        elif pos[0]<0.5 and pos[1]>0.5:
            quadrant.append(3)
        else:
            quadrant.append(4)
    return np.array(quadrant)

Gender = 'M'
L =9
#df = []
df = pd.DataFrame()
for i in range(0,L):
#   print(str(i)+'Z')
    
    dic_result = scipy.io.loadmat("Desktop/exploreSLproperties-master/results/HOMMES/experiment_"+str(i)+".mat")

    duration = dic_result["task_duration"][0]
    positionT = dic_result["stimuli_history"][:,0,2:]
    
    data = pd.DataFrame(duration)
    data['quadrantT'] = quadrant_determine(positionT)# quadrantT
    data.columns  = [ 'Duration' ,'QuadrantT']
    data["Epochs"] = data.index+1
    data  = data[['Epochs','Duration' ,'QuadrantT']]
    data['Gender']  = Gender
    data['id']  = Gender+str(i)
    df = df.append(data)
    
Gender = 'F'
L =13
#df = []
#df = pd.DataFrame()
for i in range(0,L):
#   print(str(i)+'Z')
    
    dic_result = scipy.io.loadmat("Desktop/exploreSLproperties-master/results/FEMMES/experiment_"+str(i)+".mat")

    duration = dic_result["task_duration"][0]
    positionT = dic_result["stimuli_history"][:,0,2:]
    
    data = pd.DataFrame(duration)
    data['quadrantT'] = quadrant_determine(positionT)# quadrantT
    data.columns  = [ 'Duration' ,'QuadrantT']
    data["Epochs"] = data.index+1
    data  = data[['Epochs','Duration' ,'QuadrantT']]
    data['Gender']  = Gender
    data['id']  = Gender+str(i)
    df = df.append(data)
```
